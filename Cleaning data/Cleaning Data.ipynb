```python
# Importing packages for analyzing useses
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

register_matplotlib_converters()
sns.set()
```

```python
# Import the data and making a dataframe out of it
file_name = '../Data/dataGM_excel2.xlsx'
productionDF = pd.read_excel(file_name, sheet_name='Productie')
consumptionDF = pd.read_excel(file_name, sheet_name='Consumptie')
```

```python
df_prod = productionDF.copy()
df_cons = consumptionDF.copy()

if (df_prod.index.name is None):
    df_prod = df_prod.set_index('ID-nummer')
if (df_cons.index.name is None):
    df_cons = df_cons.set_index('ID-nummer')
    
df_houses_info = df_prod[0:3]
df_prod_kwh = df_prod[4:]
df_cons_kwh = df_cons[4:]

df_prod_kwh.index = pd.to_datetime(df_prod_kwh.index)
df_cons_kwh.index = pd.to_datetime(df_cons_kwh.index)

# Columns rename
cols = df_prod_kwh.columns.values.tolist()
cols_arr_prod = []
cols_arr_cons = []

for col in cols:
    cols_arr_prod.append(col + '_prod')
    cols_arr_cons.append(col + '_cons')

df_prod_kwh.columns = cols_arr_prod
df_cons_kwh.columns = cols_arr_cons

df_prod_kwh.drop(['H05_prod', 'H10_prod', 'H12_prod', 'H14_prod', 'H30_prod'], axis=1, inplace=True)
df_cons_kwh.drop(['H05_cons', 'H10_cons', 'H12_cons', 'H14_cons', 'H30_cons'], axis=1, inplace=True)

display(df_prod_kwh.head())
display(df_cons_kwh.head())
```

```python
for house in range(1,34):
    if not ((house == 5) or (house == 10) or (house == 12) or (house == 14) or (house == 30)):
        if house < 10:
            house_no = 'H0'+ str(house)
        else:
            house_no = 'H'+ str(house)
        for i in range(len(df_prod_kwh)):
            if df_prod_kwh[house_no + '_prod'][i] > (0.25 * 0.25 * df_houses_info[house_no]['PV-aantal']):
                df_prod_kwh[house_no + '_prod'][i] = np.NaN
        print('House ' + str(house) + ' done')
```

```python
for house in range(1, 34):
    if not ((house == 5) or (house == 10) or (house == 12) or (house == 14) or (house == 30)):
        if house < 10:
            house_no = 'H0'+ str(house)
        else:
            house_no = 'H'+ str(house)
        for i in range(len(df_cons_kwh)):
            if df_cons_kwh[house_no + '_cons'][i] > (75 * 230 / 4000):
                df_cons_kwh[house_no + '_cons'][i] = np.NaN
        print('House ' + str(house) + ' done')
```

```python
df_energy_raw = pd.concat([df_prod_kwh, df_cons_kwh], axis=1)
display(df_energy_raw.head())
display(df_energy_raw.describe())
```

```python
df_energy_out = df_energy_raw.dropna(how='any', axis=0)
display(df_energy_out.head())
display(df_energy_out.describe())
```

```python
df_energy_out.to_csv('Final_data_cleaned_missing_houses.csv')
```
