```python
# Importing packages for analyzing useses
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()
sns.set()

# Import the data and making a dataframe out of it
file_name = '../Data/dataGM_excel2.xlsx'
productionDF = pd.read_excel(file_name, sheet_name='Productie')
consumptionDF = pd.read_excel(file_name, sheet_name='Consumptie')

if (productionDF.index.name is None):
    productionDF = productionDF.set_index('ID-nummer')
if (consumptionDF.index.name is None):
    consumptionDF = consumptionDF.set_index('ID-nummer')
    
df_houses_info = productionDF[0:3]
df_prod = productionDF[4:]
df_cons = consumptionDF[4:]

df_prod.index = pd.to_datetime(df_prod.index)
df_cons.index = pd.to_datetime(df_cons.index)

for column in df_prod.columns :
    combined_df = pd.merge(df_prod[column], df_cons[column], left_index=True,right_index=True)
    max_energy_prod = 0.25 * 0.25 * df_houses_info[column]['PV-aantal']
    max_energy_cons = (75 * 230) / 4000
    prod_max = combined_df[(combined_df[column + '_x'] < max_energy_prod) & (combined_df[column + '_y'] < max_energy_cons)]
    prod_max = prod_max.dropna(how='any', axis=0)
    prod_max.columns = ['production','consumption']
    globals()[str(column)] = prod_max
```
