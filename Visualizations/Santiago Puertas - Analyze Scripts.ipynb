```python
%reset
# from __future__ import division
#Importing packages for analyzing useses
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Import the data and making a dataframe out of it



file_name = '../Data/dataGM_excel2.xlsx'
productionDF = pd.read_excel(file_name, sheet_name='Productie')
consumptionDF = pd.read_excel(file_name, sheet_name='Consumptie')
productionDF.head()
consumptionDF.head()
```

```python
df_prod_1 = productionDF.iloc[4:100,0:2]
df_cons_1 = consumptionDF.iloc[4:100,1:2]

df_house_1 = df_prod_1
df_house_1.columns = ['date', 'prod']
df_house_1['cons'] = df_cons_1
df_house_1.set_index('date', inplace=True)

df_house_1.head()
```

```python
df_prod_1.time = pd.to_datetime(df_prod_1.index, format='%Y-%m-%d %H:%M:%S')
df_prod_1.plot()
plt.show()

```

```python
#PanelHouses = productionDF.iloc[1,:].str.contains('4', regex=True)
PanelHouses = productionDF.iloc[1,:] == 9
#print(PanelHouses)

PanelHouses = list(PanelHouses[PanelHouses].index.values)


print(PanelHouses)
```

```python
# x-axis values 
x = df_prod_1.time 
print(x)
# y-axis values 
y = df_house_1['cons']
print(y)
  
# setting the x - coordinates 
#x = np.arange(0, 2*(np.pi), 0.1) 
# setting the corresponding y - coordinates 
#y = np.sin(x) 
  
# potting the points 
plt.plot(x, y) 
  
# function to show the plot 
plt.show() 
```

```python
#fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,5))
#fig.suptitle('Consumption by time')
#ax1.plot(x, y)
#ax2.plot(x, -y)

df_prod_month = productionDF.iloc[4:3076,0:2]
df_cons_month = consumptionDF.iloc[4:3076,1:2]

df_house_month = df_prod_month
df_house_month.columns = ['date', 'prod']
df_house_month['cons'] = df_cons_month
df_house_month.set_index('date', inplace=True)

df_prod_month.time = pd.to_datetime(df_prod_month.index, format='%Y-%m-%d %H:%M:%S')
df_prod_month.plot()
plt.show()
```

```python
df_prod_year = productionDF.iloc[4:34180,0:2]
df_cons_year = consumptionDF.iloc[4:34180,1:2]

df_house_year = df_prod_year
df_house_year.columns = ['date', 'prod']
df_house_year['cons'] = df_cons_year
df_house_year.set_index('date', inplace=True)

df_prod_year.time = pd.to_datetime(df_prod_year.index, format='%Y-%m-%d %H:%M:%S')
#plt.ylim(0,200)
#plt.xlim(0,200)
df_prod_year.plot(alpha=0.7)
plt.ylim(0,1.5)
plt.show()
#borrar los mÃ¡ximos para ver mejor en escala los datos
#clean the maximun data to show better the scale 
```

```python
df_prod_full = productionDF.iloc[4:,0:2]
df_cons_full = consumptionDF.iloc[4:,1:2]

df_house_full = df_prod_full
df_house_full.columns = ['date', 'prod']
df_house_full['cons'] = df_cons_full
df_house_full.set_index('date', inplace=True)

df_prod_full.time = pd.to_datetime(df_prod_full.index, format='%Y-%m-%d %H:%M:%S')
df_prod_full.plot()
plt.show()
```

```python


fig, axs = plt.subplots(2, 2,figsize=(15,15))
axs[0, 0].plot(df_prod_1.time , df_house_1['cons'])
axs[0, 0].set_title('Day')

axs[0, 1].plot(df_prod_month.time, df_prod_month, 'tab:orange')
axs[0, 1].set_title('Month')

axs[1, 0].plot(df_prod_year.time, df_prod_year, 'tab:green')
axs[1, 0].set_title('Year')

axs[1, 1].plot(df_prod_full.time, df_prod_full, 'tab:red')
axs[1, 1].set_title('Full')

for ax in axs.flat:
    ax.set(xlabel='Date', ylabel='Consumption')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python


# max prod

x = data['max_prod_date']
y = data['max_prod']

# Set up the axes with gridspec
fig = plt.figure(figsize=(10, 10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=6, alpha=0.5)
main_ax.set_title('\n\nmax prod:')

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='green')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='green')
y_hist.invert_xaxis()





#________________________

# max cons


x = data['max_cons_date']
y = data['max_cons']

# Set up the axes with gridspec
fig = plt.figure(figsize=(10, 10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=6, alpha=0.5)
main_ax.set_title('\n\nmax cons:')

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='green')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='green')
y_hist.invert_xaxis()


#__________________________

#relation between solar panels and production


x = data['solar_panels']
y = data['max_prod']

# Set up the axes with gridspec
fig = plt.figure(figsize=(10, 10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=6, alpha=0.5)
main_ax.set_title('\n\nrelation between solar panels and production:')


# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='green')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='green')
y_hist.invert_xaxis()


#___________________________

#relation between people and production


x = data['people']
y = data['max_prod']

# Set up the axes with gridspec
fig = plt.figure(figsize=(10, 10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=6, alpha=0.5)
main_ax.set_title('\n\nrelation between people and production:')

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='green')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='green')
y_hist.invert_xaxis()


#_________________________________

#relation between people and cons



x = data['people']
y = data['max_cons']

# Set up the axes with gridspec
fig = plt.figure(figsize=(10, 10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=6, alpha=0.5)
main_ax.set_title('\n\nrelation between people and cons:')

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='green')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='green')
y_hist.invert_xaxis()

```
